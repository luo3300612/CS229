# Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

## Ground Truth and Parameters

```python
ground_k = 3
ground_b = 2
m = 200
non_linear_seperablility = 0.1
```

## 随机生成样本

```python
np.set_printoptions(precision=3, suppress=True)
x = np.random.uniform(-10,10,(1,m))
y = np.random.uniform(-30,30,(1,m))
```

```python
ground_x = np.linspace(-10,10,100)
ground_y = 3*ground_x + ground_b
```

```python
label_map = {"0":1,"1":0}
```

```python
label = np.zeros(x.shape)
for i in range(x.shape[1]):
    if y[0,i] - ground_k * x[0,i] - ground_b > 0:
        label[0,i] = 1
    if np.random.uniform(0,1) < non_linear_seperablility:
        label[0,i] = label_map[str(int(label[0,i]))]
```

```python
plt.scatter(x[label==0],y[label==0],marker='x')
plt.scatter(x[label==1],y[label==1],marker='o')
plt.plot(ground_x,ground_y,color='r')
```

## Get start

$$g(\eta)=\dfrac{1}{1+e^{-\eta}}$$

$$J(\theta)=-\Sigma_{i=1}^{m}\left(y^{(i)}log(g(\theta^Tx^{(i)})+(1-y^{(i)})log(1-g(\theta^Tx^{(i)}))\right)=-y^Tlog(g(X\theta))-(1-y)^Tlog(1-g(X\theta))$$

$$\dfrac{\partial J}{\partial \theta_j}=\Sigma_{i=1}^{m}\left(g(\theta^Tx^{(i)})-y^{(i)}\right)x^{(i)}_j$$

$$\dfrac{\partial J}{\partial \theta}=X^T\left(g(X\theta)-y\right)$$

```python
def sigmoid(eta):
    return 1/(1+np.e**(-eta))

def cost(Theta,X,Y):
    Y = Y.reshape((1,-1))
    pred = sigmoid(X.dot(Theta))
    return -Y.dot((np.log(sigmoid(pred))))-(1-Y).dot(np.log(1-pred))
    
def dcost(Theta,X,Y):
    Y = Y.reshape((-1,1))
    return X.T.dot(sigmoid(X.dot(Theta))-Y)
```

```python
stop_criterion = 501
alpha = 0.0001
```

```python
X_ = np.vstack((x,y))
X_ = np.vstack((np.ones((1,m)),X_)).T
```

```python
theta = np.zeros((3,1))
cost_history = np.zeros((stop_criterion,))
for epoch in range(stop_criterion):
    cost_history[epoch] = cost(theta,X_,label)
    Dcost = dcost(theta,X_,label)
    theta -= alpha * Dcost
    if epoch % 50 == 0:
        print(f"epoch:{epoch},cost:{cost_history[epoch]}")
```

```python
plt.plot(cost_history)
```

```python
x_pred = np.linspace(-10,10,100)
y_pred = (- theta[0,0] - theta[1,0]*x_pred)/theta[2,0]
```

```python
plt.plot(x_pred,y_pred,'g')
plt.plot(ground_x,ground_y,'r')
plt.scatter(x[label==0],y[label==0],marker='x')
plt.scatter(x[label==1],y[label==1],marker='o')
plt.legend(['prediction','ground truth'])
```

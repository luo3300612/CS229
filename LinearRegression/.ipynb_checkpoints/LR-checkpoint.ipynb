## 读取数据

```python
train = pd.read_csv("./all/train.csv")
test = pd.read_csv("./all/test.csv")
```

## 删除非数字的属性

```python
todrop = []
for column in train.columns:
    if train[column].dtype == 'object':
        todrop.append(column)
    #print(column,' ',data[column].dtype,' ',data[column][0])
```

```python
train = train.drop(todrop, axis=1)
test = test.drop(todrop, axis=1)
```

## 补充NULL值

```python
print(train.isnull().sum())
print('-'*50)
print(test.isnull().sum())
```

```python
datasets = [train, test]
```

```python
for dataset in datasets:
    ret = dataset.isnull().sum()
    for index in ret.index:
        if ret[index] != 0:
            dataset[index].fillna(value=dataset[index].mean(), inplace=True)
```

## 归一化

```python
train = (train - train.mean(0))/train.std(0)
test = (test - test.mean(0))/test.std(0)
```

```python
train = np.hstack((np.ones((train.shape[0],1)),train))
test = np.hstack((np.ones((test.shape[0],1)),test))
```

```python
origin_train = train.copy()
train = train[:1000, :]
```

```python
def cal_cost(theta, x, y):
    return 0.5 * ((x.reshape(1,-1)).dot(theta) - y)**2

def numerical_gradient(theta, x, y):
    dtheta = np.zeros((theta.shape))
    n = theta.shape[0]
    for i in range(n):
        theta[i] += 0.01
        cost_up = cal_cost(theta, x, y)
        theta[i] -= 0.02
        cost_down = cal_cost(theta, x, y)
        dtheta[i] = (cost_up - cost_down) / 0.02
    return dtheta
```

```python
train = origin_train
theta = np.zeros((37,1))
itern = 1000
m = train.shape[0]
n = train.shape[1]
alpha = 0.0001
cost_history = np.zeros((itern,))
for epoch in range(itern):
    for i in range(m):
        x = train[i,:].reshape(-1,1)
        y = label[i]
        cost = cal_cost(theta, x, y)
        dtheta = np.zeros((theta.shape))
        
        x_dot_theta = (x.reshape(1,-1)).dot(theta)
        dtheta[0,0] = x_dot_theta - y
        for j in range(1,dtheta.shape[0]):
            dtheta[j] = (x_dot_theta - y) * x[j,0]
            #print(dtheta[j])
        #print('+'*40)
        #ng = numerical_gradient(theta, x, y)
        #print(np.hstack((dtheta,ng)))
        theta = theta - alpha * dtheta
        
        cost_history[epoch] += cost
        #print('epoch:%d,cost:%.2f' % (epoch, cost))
    cost_history[epoch] /= m
    #print(dtheta)
    print('epoch_cost:%.2f' % (cost_history[epoch]))
```

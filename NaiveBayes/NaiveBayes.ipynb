# 朴素贝叶斯

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.DataFrame(columns=['色泽','根蒂','敲声','纹理','脐部','触感'])
```

```python
data['色泽'] = ['青绿','乌黑','乌黑','青绿','浅白','青绿','乌黑','乌黑','乌黑','青绿','浅白','浅白','青绿','浅白','乌黑','浅白','青绿']
data['根蒂'] = ['蜷缩','蜷缩','蜷缩','蜷缩','蜷缩','稍蜷','稍蜷','稍蜷','稍蜷','硬挺','硬挺','蜷缩','稍蜷','稍蜷','稍蜷','蜷缩','蜷缩']
data['敲声'] = ['浊响','沉闷','浊响','沉闷','浊响','浊响','浊响','浊响','沉闷','清脆','清脆','浊响','浊响','沉闷','浊响','浊响','沉闷']
data['纹理'] = ['清晰','清晰','清晰','清晰','清晰','清晰','稍糊','清晰','稍糊','清晰','模糊','模糊','稍糊','稍糊','清晰','模糊','稍糊']
data['脐部'] = ['凹陷','凹陷','凹陷','凹陷','凹陷','稍凹','稍凹','稍凹','稍凹','平坦','平坦','平坦','凹陷','凹陷','稍凹','平坦','稍凹']
data['触感'] = ['硬滑','硬滑','硬滑','硬滑','硬滑','软粘','软粘','硬滑','硬滑','软粘','硬滑','软粘','硬滑','硬滑','软粘','硬滑','硬滑']
data['好瓜'] = ['是','是','是','是','是','是','是','是','否','否','否','否','否','否','否','否','否']
```

```python
data
```

```python
train = pd.concat([data[0:6],data[8:15]])
test = pd.concat([data[6:8],data[15:17]])
```

```python
train
```

```python
test
```

```python
def cal_condition_probability(data,column,attr,label):
    target_label_data = data[data['好瓜']==label]
    target_label_and_attr_data = target_label_data[target_label_data[column] == attr]
    return target_label_and_attr_data.shape[0]/target_label_data.shape[0]
```

```python
column = test.columns
```

```python
def cal_prob(to_pred,label):
    p = train[train['好瓜']==label].shape[0]/train.shape[0]
    for i in range(len(column)-1):
        p *= cal_condition_probability(train,column[i],to_pred[column[i]],label)
    return p
```

```python
for i in range(test.shape[0]):
    print('瓜：')
    print(test.iloc[i,:].values)
    print('好瓜得分：',cal_prob(test.iloc[i,:],'是'))
    print('非好瓜得分：',cal_prob(test.iloc[i,:],'否'))
    print('判断结果：',end='')
    if cal_prob(test.iloc[i,:],'是') > cal_prob(test.iloc[i,:],'否'):
        print('好瓜')
    else:
        print('非好瓜')
    print()
```

### 拉普拉斯平滑

```python
def cal_prob_with_laplace_smoothing(to_pred,label):
    #print('p = ','')
    #print(train[train['好瓜']==label].shape[0]+1)
    #print(train.shape[0]+train['好瓜'].value_counts().shape[0])
    p = (train[train['好瓜']==label].shape[0]+1)/(train.shape[0]+train['好瓜'].value_counts().shape[0])
    #print(p,'* ',end='')
    for i in range(len(column)-1):
        to_mul = cal_condition_probability_with_laplace_smoothing(train,column[i],to_pred[column[i]],label)
        p *= to_mul
        #print(to_mul,'* ',end='')
    #print()
    return p

def cal_condition_probability_with_laplace_smoothing(data,column,attr,label):
    target_label_data = data[data['好瓜']==label]
    target_label_and_attr_data = target_label_data[target_label_data[column] == attr]
    #print()
    #print(target_label_and_attr_data)
    #print(target_label_and_attr_data.shape[0]+1)
    #print(target_label_data.shape[0]+target_label_data[column].value_counts().shape[0])
    return (target_label_and_attr_data.shape[0]+1)/(target_label_data.shape[0]+target_label_data[column].value_counts().shape[0])
```

```python
for i in range(test.shape[0]):
    print('瓜：')
    print(test.iloc[i,:].values)
    score_good = cal_prob_with_laplace_smoothing(test.iloc[i,:],'是')
    score_bad = cal_prob_with_laplace_smoothing(test.iloc[i,:],'否')
    
    print('好瓜得分：',score_good)
    print('非好瓜得分：',score_bad)
    print('判断结果：',end='')
    if score_good > score_bad:
        print('好瓜')
    else:
        print('非好瓜')
    print()
```
